#### CLEANING DATA

Membersihkan data df_day

```python
df_hour.drop('instant', axis = 1, inplace = True)
df_day.drop('instant', axis = 1, inplace = True)
```

```python
df_hour.head()
```

```python
#Mengganti nama columns
rename = {'dteday': 'date', 'yr': 'year', 'mnth': 'month', 'hr': 'hour', 'hum': 'humidity', 'cnt': 'count'}

df_hour.rename(columns=rename, inplace=True)
df_day.rename(columns=rename, inplace=True)
```

```python
df_hour.head()
```

```python
categorical = ['season', 'month', 'weekday', 'weathersit', 'holiday', 'workingday']

for df in [df_day, df_hour]:
    df['date'] = pd.to_datetime(df['date'])
    df[categorical] = df[categorical].astype('category')
    
df_day.info()
```

```python
#Labelling
season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
for df in [df_day, df_hour]:
    df['season'].replace(season_mapping, inplace = True)
    
df_day.head()
```

```python
#Mengganti Value
year_mapping = {0: 2011, 1: 2012}
for df in [df_day, df_hour]:
    df['year'].replace(year_mapping, inplace = True)
    
df_day.head()
```

```python
#Labelling
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
for df in [df_day, df_hour]:
    df['month'].replace(month_mapping, inplace = True)
    
df_day.head()
```

```python
#Labelling
holiday_mapping = {0: 'No', 1: 'Yes'}
for df in[df_day, df_hour]:
    df['holiday'].replace(holiday_mapping, inplace = True)
    
df_day.head()
```

```python
#Labelling
weekday_mapping = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'}
for df in [df_day, df_hour]:
    df['weekday'].replace(weekday_mapping, inplace = True)
    
df_day.head()
```

```python
#Labelling
workingday_mapping = {0: 'No', 1: 'Yes'}
for df in[df_day, df_hour]:
    df['workingday'].replace(workingday_mapping, inplace = True)
    
df_day.head()
```

```python
#Labelling
weathersit_mapping = {1: 'Clear', 2: 'Mist', 3: 'Light Rain', 4: 'Heavy Rain'}
for df in [df_day, df_hour]:
    df['weathersit'].replace(weathersit_mapping, inplace = True)
    
df_hour.head()
```

```python
#Normalized Value
scaling_columns = ['temp', 'atemp', 'humidity', 'windspeed']
scaling_factors = {'temp': 41, 'atemp': 50, 'humidity': 100, 'windspeed': 67}

for df in [df_day, df_hour]:
    df[scaling_columns] = df[scaling_columns].multiply(scaling_factors)

df_hour.head()
```

#### EXPLANATORY ANALYSIS

##### Apa jam dan hari tersibuk untuk bike sharing dalam beberapa tahun terakhir?

```python
custom_palette = sns.color_palette(['#e60000', '#1919ff'])

plt.figure(figsize = (10, 5))
sns.barplot(x = 'hour', y = 'count', data = df_hour, hue = 'year', errorbar=None, palette=custom_palette)
plt.title('Jumlah pengguna bike sharing berdasarkan jam dan tahun')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.barplot(x = 'weekday', y = 'count', data = df_hour, hue = 'year',  errorbar=None, palette=custom_palette)
plt.title('Jumlah pengguna bike sharing berdasarkan hari dan tahun')
plt.show()
```

##### Bagaimana perfoma penggunaan bike sharing dalam beberapa tahun terakhir?

```python
df_2012 = df_day[df_day['year'] == 2012]
df_2011 = df_day[df_day['year'] == 2011]

monthly_df_2012 = df_2012.groupby(by='month').agg({
    "count": "sum"
})
monthly_df_2011 = df_2011.groupby(by='month').agg({
    "count": "sum"
})

monthly_df_2012 = monthly_df_2012.reset_index()
monthly_df_2011 = monthly_df_2011.reset_index()

plt.figure(figsize=(12, 5))
plt.plot(monthly_df_2012['month'], monthly_df_2012['count'], label="2012", marker=".",color='#e60000', linewidth=2)
plt.plot(monthly_df_2011['month'], monthly_df_2011['count'], label="2011", marker=".",color='#1919ff', linewidth=2)
plt.legend()
plt.show()
```

##### Musim Apa Dengan Penggunaan Bike Share Tertinggi?

```python
colors_ = ['#d1d1ff',  '#d1d1ff',"#1919ff", "#d1d1ff"]

plt.figure(figsize = (10, 5))
sns.barplot(x = 'season', y = 'count', data = df_day.sort_values(by='season', ascending=False), palette=colors_,  errorbar=None)
plt.title('Jumlah pengguna bike sharing berdasarkan musim')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```
